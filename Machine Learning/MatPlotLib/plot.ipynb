```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
```

```python
plt.plot(x,y,'r--P',label='y=2x')
plt.legend()
plt.title('test')
plt.xlabel('x')
plt.ylabel('y')
```

```python
x1=[1,2,3,4,5]
y1=[1,4,9,16,25]
```

```python
plt.plot(x1,y1,'b:D' , label='y=x^2')
plt.legend()
```

```python
plt.plot(x,y,'r--P',label='y=2x')
plt.plot(x1,y1,'b:D' , label='y=x^2')
```

```python
plt.subplot(211)
plt.plot(x,y,'r--P',label='y=2x')
plt.legend()

plt.subplot(212)
plt.plot(x1,y1,'b:D' , label='y=x^2')
plt.legend()

plt.savefig('plot.png')
```
