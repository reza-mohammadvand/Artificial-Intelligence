```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Titanic dataset
data = pd.read_csv('titanic.csv')
```

```python
# Displaying first few rows of the dataset
data.head()
```

```python
# Histogram of 'Age' column
data['Age'].hist()
```

```python
# Histogram of 'Age' column with specified number of bins
data['Age'].hist(bins=40)
```

```python
# Histogram of 'SibSp' column with specified number of bins
data['SibSp'].hist(bins=8)
```

```python
# Histogram of 'Pclass' column with specified number of bins
data['Pclass'].hist(bins=20)
```

```python
# Histogram of 'Survived' column with specified number of bins
data['Survived'].hist(bins=2)
```

```python
# Setting seaborn style to 'darkgrid'
sns.set_style('darkgrid')
```

```python
# Histogram of 'Survived' column with specified number of bins after setting seaborn style
data['Survived'].hist(bins=2)
```

```python
# Histogram of 'Survived' column using plot function
data['Survived'].plot(kind='hist', bins=2)
```

```python
# Displaying first few rows of the dataset
data.head()
```

```python
# Line plot of 'Pclass' against 'Age' after sorting values by 'Age'
data.sort_values(['Age']).plot.line(x='Age', y='Pclass', figsize=(10, 8))
```

```python
# Line plot of 'Fare' against 'Age' after sorting values by 'Age'
data.sort_values(['Age']).plot.line(x='Age', y='Fare', figsize=(10, 8))
```

```python
# Line plot of 'Fare' against 'Embarked' after sorting values by 'Embarked'
data.sort_values(['Embarked']).plot.line(x='Embarked', y='Fare', figsize=(10, 8))
```

```python
# Scatter plot of 'Age' against 'Pclass'
data.plot.scatter(x='Age', y='Pclass', figsize=(10, 8))
```

```python
# Scatter plot of 'Age' against 'Fare'
data.plot.scatter(x='Age', y='Fare', figsize=(10, 8))
```

```python
# Hexbin plot of 'Age' against 'Fare'
data.plot.hexbin(x='Age', y='Fare', gridsize=10, figsize=(10, 8))
```

```python
# Hexbin plot of 'Age' against 'Pclass'
data.plot.hexbin(x='Age', y='Pclass', gridsize=10, figsize=(10, 8))
```

```python
# Displaying first few rows of the dataset
data.head()
```

```python
# Kernel density estimation plot of 'Age'
data['Age'].plot.kde()
```

```python
# Kernel density estimation plot of 'Fare'
data['Fare'].plot.kde()
```

```python
# Histogram of all columns in the dataset
data.hist(figsize=(10, 12))
```

```python
# Box plot of all numerical columns in the dataset except 'PassengerId', 'Fare', and 'Age'
data.drop(['PassengerId', 'Fare', 'Age'], axis=1).plot.box(figsize=(12, 8))
```
