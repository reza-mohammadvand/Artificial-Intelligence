```python
import numpy as np
```

```python
q_table = np.zeros([env.observation_space.n, env.action_space.n]) # q-table full 0 ([tedad satr, tedad soton])
q_table
```

```python
import random # we can use algorith from librabry but below we type it
alpha = 0.1 # alpha
gamma = 0.6 # landa
epsilon = 0.1
all_epochs = []
all_penalties = []
for i in range(1, 10000): # 10000 times do ( a full cycle of algorithm)
    state = env.reset() # reset enviroment
    epochs, penalties, reward = 0, 0, 0
    done = False
    while not done:
        if random.uniform(0, 1) < epsilon: # اگر عدد رندوم کمتر از 0.1 بود بیا تصادفی اکشن بساز
            action = env.action_space.sample()
        else: # در غیر این صورت ایندکس بیشترین عنصر درون سطر مربوطه در جدول کیو را به ما بده
            action = np.argmax(q_table[state])
        next_state, reward, done, _ = env.step(action) # add info to variables 
        old_value = q_table[state, action] # old value of q in q-table
        next_max = np.max(q_table[next_state]) # max q that we have in next state
        new_value = (1 - alpha) * old_value + alpha * (reward + gamma * next_max) # furmol
        q_table[state, action] = new_value  # update q-table
        
        if reward == -10:
            penalties += 1
        state = next_state
        epochs += 1
    
    if i % 10 == 0: # هر 10 اپیزود (دور)
        clear_output(wait=True)
        print('Episode: ', i)
        print('Penalties: ', penalties)
        print('Epochs: ', epochs)

# در خروجی زیر میفهمیم که در آن دور با چند حرکت (ایپاک) مسافر را به مقصد رسانده ایم و با چه پنالتی
```
